```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
from pandas import ExcelFile
import scipy
from scipy import stats
import sys

desired_cols =list(range(0,12))
df = pd.read_excel('data.xlsx', sheet_name = 'CPD I WT', usecols=desired_cols)
df.dropna(inplace=True)


kd = df.iloc[:,1] 
affinity_1 = df.iloc[:,2]
affinity_2 = df.iloc[:,3]
affinity_3 = df.iloc[:,4]
Fe1 = df.iloc[:,5]
Fe2 = df.iloc[:,6]
Fe3 = df.iloc[:,7]
O1 = df.iloc[:,8]
O2 = df.iloc[:,9]
O3 = df.iloc[:,10]
```

```python
# Kd v affinity

kd_aff1 = plt.scatter(kd,affinity_1, marker='.')
kd_aff2 = plt.scatter(kd, affinity_2,marker='.')
kd_aff3 = plt.scatter(kd, affinity_3, marker='.')
kd_aff2 = plt.xlabel('WT K$_{D}$')
kd_aff2 = plt.ylabel('Affinity')
plt.legend()
```

```python
#Kd v Fe distance

kd_fe1 = plt.scatter(kd, Fe1, marker = '.')
kd_fe2 = plt.scatter(kd, Fe2, marker = '.')
kd_fe3 = plt.scatter(kd, Fe3, marker = '.')
kd_fe1 = plt.xlabel('WT K$_{D}$')
kd_fe1 = plt.ylabel('Distance to closest atom to Fe')
plt.legend()
```

```python
#affinity v Fe distance

aff_fe1 = plt.scatter(affinity_1, Fe1, marker = '.')
aff_fe2 = plt.scatter(affinity_2, Fe2, marker = '.')
aff_fe3 = plt.scatter(affinity_3, Fe3, marker = '.')
aff_fe1 = plt.xlabel('Affinity')
aff_fe1 = plt.ylabel('Distance to closest atom to Fe')
plt.legend()
```

```python
#Affinity v O distance 

aff_o1 = plt.scatter(affinity_1, O1, marker = '.')
aff_o2 = plt.scatter(affinity_2, O2, marker = '.')
aff_o3 = plt.scatter(affinity_3, O3, marker = '.')
aff_o1 = plt.xlabel('Affinity')
aff_o1 = plt.ylabel('Distance to closest atom to O')
plt.legend()
```

```python
#Kd v O distance

kd_o1 = plt.scatter(kd, O1, marker = '.')
kd_o2 = plt.scatter(kd, O2, marker = '.')
kd_o3 = plt.scatter(kd, O3, marker = '.')
kd_o1 = plt.xlabel('WT K$_{D}$')
kd_o1 = plt.ylabel('Distance to closest atom to O')
plt.legend()
```

```python
#Pearson -- Kd/affinity

kd_affinity_1 = stats.pearsonr(kd, affinity_1)
print("The Pearson's correlation coefficient for KD and the first vina affinity score is:", kd_affinity_1)

kd_affinity_2 = stats.pearsonr(kd, affinity_2)
print("The Pearson's correlation coefficient for KD and the second vina affinity score is:",kd_affinity_2)

kd_affinity_3 = stats.pearsonr(kd, affinity_3)
print("The Pearson's correlation coefficient for KD and the third vina affinity score is:",kd_affinity_3)
```

```python
#Pearson -- affinity/Fe distance

affinity1_fe1 = stats.pearsonr(affinity_1,Fe1)
print("The Pearson's correlation for the first affinity and distance is: ", affinity1_fe1)

affinity2_fe2 = stats.pearsonr(affinity_2,Fe2)
print("The Pearson's correlation for the second affinity and distance is: ", affinity2_fe2)

affinity3_fe3 = stats.pearsonr(affinity_3,Fe3)
print("The Pearson's correlation for the third affinity and distance is: ", affinity3_fe3)
```

```python
#Pearson -- affinity/O distance

affinity_o1 = stats.pearsonr(affinity_1, O1)
print("The Pearson's correlation correlation coefficient for affinity and the closest atom to the axial O is:", affinity_o1)
affinity_o2 = stats.pearsonr(affinity_2, O2)
print("The Pearson's correlation correlation coefficient for affinity and the closest atom to the axial O is:", affinity_o2)
affinity_o3 = stats.pearsonr(affinity_3, O3)
print("The Pearson's correlation correlation coefficient for affinity and the closest atom to the axial O is:", affinity_o3)
```

```python
#Pearson -- KD/ O distance

kd_o1 = stats.pearsonr(kd,O1)
print("The Pearson's correlation for the KD and distance to the closest atom from the O is: ", kd_o1)

kd_o2 = stats.pearsonr(kd, O2)
print("The Pearson's correlation for the KD and distance to the closest atom from the O is: ", kd_o2)

kd_o3 = stats.pearsonr(kd, O3)
print("The Pearson's correlation for the KD and distance to the closest atom from the O is: ", kd_o3)
```

```python
#Pearson -- Kd/Fe distance 

kd_fe_1 = stats.pearsonr(kd,Fe1)
print("The Pearson's correlation coefficient for Kd and the first distance is:", kd_fe_1)

kd_fe_2 = stats.pearsonr(kd,Fe2)
print("The Pearson's correlation coefficient for Kd and the second distance is:", kd_fe_2)

kd_fe_3 = stats.pearsonr(kd,Fe3)
print("The Pearson's correlation coefficient for Kd and the third distance is:", kd_fe_3)

```
