```python
#Import all relevant/useful modules

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
from pandas import ExcelFile
import scipy
from scipy import stats
import sys

#Read in Excel file and read desired columns, drop missing values

desired_cols =list(range(0,16))
df = pd.read_excel('data.xlsx', sheet_name = 'Rigid WT', usecols=desired_cols)
df.dropna(inplace=True)

#Assign columns to variables

wt_kd = df.iloc[:,0]
wt_affinity1 = df.iloc[:,1]
wt_affinity2 = df.iloc[:,2]
wt_affinity3 = df.iloc[:,3]
wt_dist1 = df.iloc[:,4]
wt_dist2 = df.iloc[:,5]
wt_dist3 = df.iloc[:,6]
cpd_I_WT_affinity1 = df.iloc[:,7]
cpd_I_WT_affinity2 = df.iloc[:,8]
cpd_I_WT_affinity3 = df.iloc[:,9]
cpd_I_WT_O_distance1 = df.iloc[:,10]
cpd_I_WT_O_distance2 = df.iloc[:,11]
cpd_I_WT_O_distance3 = df.iloc[:,12]
cpd_I_WT_Fe_distance1 = df.iloc[:,13]
cpd_I_WT_Fe_distance2 = df.iloc[:,14]
cpd_I_WT_Fe_distance3 = df.iloc[:,15]
```

```python
# WT KD v Affinity

kd_aff = plt.plot(wt_kd, wt_affinity1, marker='.', linestyle='none')
kd_aff = plt.plot(wt_kd, wt_affinity2, marker='.', linestyle='none')
kd_aff = plt.plot(wt_kd, wt_affinity3, marker='.', linestyle='none')
kd_aff = plt.xlabel('WT K$_{D}$ / kcal mol$^{-1}$')
kd_aff = plt.ylabel('Affinity / kcal mol$^{-1}$')
plt.legend()

print(stats.pearsonr(wt_kd, wt_affinity1))
print(stats.pearsonr(wt_kd, wt_affinity2))
print(stats.pearsonr(wt_kd, wt_affinity3))
```

```python
# WT KD v CPD I WT Affinity

kd_aff_cpd_I = plt.plot(wt_kd, cpd_I_WT_affinity1, marker='.', linestyle='none')
kd_aff_cpd_I = plt.plot(wt_kd, cpd_I_WT_affinity2, marker='.', linestyle='none')
kd_aff_cpd_I = plt.plot(wt_kd, cpd_I_WT_affinity3, marker='.', linestyle='none')
kd_aff_cpd_I = plt.xlabel('WT K$_{D}$ / kcal mol$^{-1}$')
kd_aff_cpd_I = plt.ylabel('Compound I WT Affinity / kcal mol$^{1}$')
plt.legend()

print(stats.pearsonr(wt_kd, cpd_I_WT_affinity1))
print(stats.pearsonr(wt_kd, cpd_I_WT_affinity2))
print(stats.pearsonr(wt_kd, cpd_I_WT_affinity3))
```

```python
# WT KD v Closest atom

kd_wt_dist = plt.plot(wt_kd,wt_dist1, marker ='.', linestyle='none')
kd_wt_dist = plt.plot(wt_kd,wt_dist2, marker ='.', linestyle='none')
kd_wt_dist = plt.plot(wt_kd,wt_dist3, marker ='.', linestyle='none')
kd_wt_dist = plt.xlabel('WT K$_{D}$ / kcal mol$^{-1}$')
kd_wt_dist = plt.ylabel('Distance / $\mathring{A}$')
plt.legend()

print(stats.pearsonr(wt_kd, wt_dist1))
print(stats.pearsonr(wt_kd, wt_dist2))
print(stats.pearsonr(wt_kd, wt_dist3))
```

```python
# WT KD v CPD I WT Closest atom from O 

kd_cpd_I_wt_dist = plt.plot(wt_kd,cpd_I_WT_O_distance1, marker ='.', linestyle='none')
kd_cpd_I_wt_dist = plt.plot(wt_kd,cpd_I_WT_O_distance2, marker ='.', linestyle='none')
kd_cpd_I_wt_dist = plt.plot(wt_kd,cpd_I_WT_O_distance3, marker ='.', linestyle='none')
kd_cpd_I_wt_dist = plt.xlabel('WT K$_{D}$ / kcal mol$^{-1}$')
kd_cpd_I_wt_dist = plt.ylabel('Distance from Compound I WT O / $\mathring{A}$')
plt.legend()

print(stats.pearsonr(wt_kd, cpd_I_WT_O_distance1))
print(stats.pearsonr(wt_kd, cpd_I_WT_O_distance2))
print(stats.pearsonr(wt_kd, cpd_I_WT_O_distance3))
```

```python
# WT KD v CPD I WT Closest atom from Fe 

kd_cpd_I_wt_dist_fe = plt.plot(wt_kd,cpd_I_WT_Fe_distance1, marker ='.', linestyle='none')
kd_cpd_I_wt_dist_fe = plt.plot(wt_kd,cpd_I_WT_Fe_distance2, marker ='.', linestyle='none')
kd_cpd_I_wt_dist_fe = plt.plot(wt_kd,cpd_I_WT_Fe_distance3, marker ='.', linestyle='none')
kd_cpd_I_wt_dist_fe = plt.xlabel('WT K$_{D}$ / kcal mol $^{-1}$')
kd_cpd_I_wt_dist_fe = plt.ylabel('Distance from Compound I WT Fe / $\mathring{A}$')
plt.legend()

print(stats.pearsonr(wt_kd, cpd_I_WT_Fe_distance1))
print(stats.pearsonr(wt_kd, cpd_I_WT_Fe_distance2))
print(stats.pearsonr(wt_kd, cpd_I_WT_Fe_distance3))
```

```python
# WT Affinity v distance

wt_aff_dist = plt.plot(wt_affinity1, wt_dist1, marker='.', linestyle='none')
wt_aff_dist = plt.plot(wt_affinity2, wt_dist2, marker='.', linestyle='none')
wt_aff_dist = plt.plot(wt_affinity3, wt_dist3, marker='.', linestyle='none')
wt_aff_dist = plt.xlabel('WT affinity / kcal mol $^{-1}$')
wt_aff_dist = plt.ylabel('Distance to closest ligand atom / $\mathring{A}$')
plt.legend()

print(stats.pearsonr(wt_affinity1, wt_dist1))
print(stats.pearsonr(wt_affinity2, wt_dist2))
print(stats.pearsonr(wt_affinity3, wt_dist3))
```

```python
#CPD I WT Affinity v CPD I WT Distance to Fe

cpd_i_wt_aff_fe = plt.scatter(cpd_I_WT_affinity1,cpd_I_WT_Fe_distance1, marker='.')
cpd_i_wt_aff_fe = plt.scatter(cpd_I_WT_affinity2,cpd_I_WT_Fe_distance2, marker='.')
cpd_i_wt_aff_fe = plt.scatter(cpd_I_WT_affinity3,cpd_I_WT_Fe_distance3, marker='.')
cpd_i_wt_aff_fe = plt.xlabel('Compound I WT affinity / kcal mol$^{-1}$')
cpd_i_wt_aff_fe = plt.ylabel('Closest distance from central Iron / $\mathring{A}$')
plt.legend()

print(stats.pearsonr(cpd_I_WT_affinity1, cpd_I_WT_Fe_distance1))
print(stats.pearsonr(cpd_I_WT_affinity2, cpd_I_WT_Fe_distance2))
print(stats.pearsonr(cpd_I_WT_affinity3, cpd_I_WT_Fe_distance3))
```

```python
#CPD I WT Affinity v CPD I WT Distance to O

cpd_i_wt_aff_o = plt.scatter(cpd_I_WT_affinity1,cpd_I_WT_Fe_distance1, marker='.')
cpd_i_wt_aff_o = plt.scatter(cpd_I_WT_affinity2,cpd_I_WT_Fe_distance3, marker='.')
cpd_i_wt_aff_o = plt.scatter(cpd_I_WT_affinity3,cpd_I_WT_Fe_distance3, marker='.')
cpd_i_wt_aff_o = plt.xlabel('Compound I WT affinity / kcal mol $^{-1}$')
cpd_i_wt_aff_o = plt.ylabel('Closest distance from central Oxygen / $\mathring{A}')
plt.legend()

print(stats.pearsonr(cpd_I_WT_affinity1, cpd_I_WT_Fe_distance1))
print(stats.pearsonr(cpd_I_WT_affinity2, cpd_I_WT_Fe_distance2))
print(stats.pearsonr(cpd_I_WT_affinity3, cpd_I_WT_Fe_distance3))
```
