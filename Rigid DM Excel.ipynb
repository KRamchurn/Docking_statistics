```python
#Import all relevant/useful modules

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
from pandas import ExcelFile
import scipy
from scipy import stats
import sys

#Read in Excel file and read desired columns, drop missing values

desired_cols =list(range(0,16))
df = pd.read_excel('data.xlsx', sheet_name = 'Rigid DM', usecols=desired_cols)
df.dropna(inplace=True)

#Assign columns to variables

dm_kd = df.iloc[:,0]
dm_affinity1 = df.iloc[:,1]
dm_affinity2 = df.iloc[:,2]
dm_affinity3 = df.iloc[:,3]
dm_dist1 = df.iloc[:,4]
dm_dist2 = df.iloc[:,5]
dm_dist3 = df.iloc[:,6]
cpd_I_dm_affinity1 = df.iloc[:,7]
cpd_I_dm_affinity2 = df.iloc[:,8]
cpd_I_dm_affinity3 = df.iloc[:,9]
cpd_I_dm_O_distance1 = df.iloc[:,10]
cpd_I_dm_O_distance2 = df.iloc[:,11]
cpd_I_dm_O_distance3 = df.iloc[:,12]
cpd_I_dm_Fe_distance1 = df.iloc[:,13]
cpd_I_dm_Fe_distance2 = df.iloc[:,14]
cpd_I_dm_Fe_distance3 = df.iloc[:,15]
```

```python
# dm KD v Affinity

kd_aff = plt.plot(dm_kd, dm_affinity1, marker='.', linestyle='none')
kd_aff = plt.plot(dm_kd, dm_affinity2, marker='.', linestyle='none')
kd_aff = plt.plot(dm_kd, dm_affinity3, marker='.', linestyle='none')
kd_aff = plt.xlabel('DM K$_{D}$ / kcal mol$^{-1}$')
kd_aff = plt.ylabel('Affinity / kcal mol$^{-1}$')
plt.legend()

print(stats.pearsonr(dm_kd, dm_affinity1))
print(stats.pearsonr(dm_kd, dm_affinity2))
print(stats.pearsonr(dm_kd, dm_affinity3))
```

```python
# dm KD v CPD I dm Affinity

kd_aff_cpd_I = plt.plot(dm_kd, cpd_I_dm_affinity1, marker='.', linestyle='none')
kd_aff_cpd_I = plt.plot(dm_kd, cpd_I_dm_affinity2, marker='.', linestyle='none')
kd_aff_cpd_I = plt.plot(dm_kd, cpd_I_dm_affinity3, marker='.', linestyle='none')
kd_aff_cpd_I = plt.xlabel('DM K$_{D}$ / kcal mol$^{-1}$')
kd_aff_cpd_I = plt.ylabel('Compound I DM Affinity / kcal mol$^{1}$')
plt.legend()

print(stats.pearsonr(dm_kd, cpd_I_dm_affinity1))
print(stats.pearsonr(dm_kd, cpd_I_dm_affinity2))
print(stats.pearsonr(dm_kd, cpd_I_dm_affinity3))
```

```python
# dm KD v Closest atom

kd_dm_dist = plt.plot(dm_kd,dm_dist1, marker ='.', linestyle='none')
kd_dm_dist = plt.plot(dm_kd,dm_dist2, marker ='.', linestyle='none')
kd_dm_dist = plt.plot(dm_kd,dm_dist3, marker ='.', linestyle='none')
kd_dm_dist = plt.xlabel('DM K$_{D}$ / kcal mol$^{-1}$')
kd_dm_dist = plt.ylabel('Distance / $\mathring{A}$')
plt.legend()

print(stats.pearsonr(dm_kd, dm_dist1))
print(stats.pearsonr(dm_kd, dm_dist2))
print(stats.pearsonr(dm_kd, dm_dist3))
```

```python
# dm KD v CPD I dm Closest atom from O 

kd_cpd_I_dm_dist = plt.plot(dm_kd,cpd_I_dm_O_distance1, marker ='.', linestyle='none')
kd_cpd_I_dm_dist = plt.plot(dm_kd,cpd_I_dm_O_distance2, marker ='.', linestyle='none')
kd_cpd_I_dm_dist = plt.plot(dm_kd,cpd_I_dm_O_distance3, marker ='.', linestyle='none')
kd_cpd_I_dm_dist = plt.xlabel('DM K$_{D}$ / kcal mol$^{-1}$')
kd_cpd_I_dm_dist = plt.ylabel('Distance from Compound I DM O / $\mathring{A}$')
plt.legend()

print(stats.pearsonr(dm_kd, cpd_I_dm_O_distance1))
print(stats.pearsonr(dm_kd, cpd_I_dm_O_distance2))
print(stats.pearsonr(dm_kd, cpd_I_dm_O_distance3))
```

```python
# dm KD v CPD I dm Closest atom from Fe 

kd_cpd_I_dm_dist_fe = plt.plot(dm_kd,cpd_I_dm_Fe_distance1, marker ='.', linestyle='none')
kd_cpd_I_dm_dist_fe = plt.plot(dm_kd,cpd_I_dm_Fe_distance2, marker ='.', linestyle='none')
kd_cpd_I_dm_dist_fe = plt.plot(dm_kd,cpd_I_dm_Fe_distance3, marker ='.', linestyle='none')
kd_cpd_I_dm_dist_fe = plt.xlabel('DM K$_{D}$ / kcal mol $^{-1}$')
kd_cpd_I_dm_dist_fe = plt.ylabel('Distance from Compound I DM Fe / $\mathring{A}$')
plt.legend()

print(stats.pearsonr(dm_kd, cpd_I_dm_Fe_distance1))
print(stats.pearsonr(dm_kd, cpd_I_dm_Fe_distance2))
print(stats.pearsonr(dm_kd, cpd_I_dm_Fe_distance3))
```

```python
# dm Affinity v distance

dm_aff_dist = plt.plot(dm_affinity1, dm_dist1, marker='.', linestyle='none')
dm_aff_dist = plt.plot(dm_affinity2, dm_dist2, marker='.', linestyle='none')
dm_aff_dist = plt.plot(dm_affinity3, dm_dist3, marker='.', linestyle='none')
dm_aff_dist = plt.xlabel('DM affinity / kcal mol $^{-1}$')
dm_aff_dist = plt.ylabel('Distance to closest ligand atom / $\mathring{A}$')
plt.legend()

print(stats.pearsonr(dm_affinity1, dm_dist1))
print(stats.pearsonr(dm_affinity2, dm_dist2))
print(stats.pearsonr(dm_affinity3, dm_dist3))
```

```python
#CPD I dm Affinity v CPD I dm Distance to Fe

cpd_i_dm_aff_fe = plt.scatter(cpd_I_dm_affinity1,cpd_I_dm_Fe_distance1, marker='.')
cpd_i_dm_aff_fe = plt.scatter(cpd_I_dm_affinity2,cpd_I_dm_Fe_distance2, marker='.')
cpd_i_dm_aff_fe = plt.scatter(cpd_I_dm_affinity3,cpd_I_dm_Fe_distance3, marker='.')
cpd_i_dm_aff_fe = plt.xlabel('Compound I DM affinity / kcal mol$^{-1}$')
cpd_i_dm_aff_fe = plt.ylabel('Closest distance from Fe / $\mathring{A}$')
plt.legend()

print(stats.pearsonr(cpd_I_dm_affinity1, cpd_I_dm_Fe_distance1))
print(stats.pearsonr(cpd_I_dm_affinity2, cpd_I_dm_Fe_distance2))
print(stats.pearsonr(cpd_I_dm_affinity3, cpd_I_dm_Fe_distance3))
```

```python
#CPD I dm Affinity v CPD I dm Distance to O

cpd_i_dm_aff_o = plt.scatter(cpd_I_dm_affinity1,cpd_I_dm_O_distance1, marker='.')
cpd_i_dm_aff_o = plt.scatter(cpd_I_dm_affinity2,cpd_I_dm_O_distance2, marker='.')
cpd_i_dm_aff_o = plt.scatter(cpd_I_dm_affinity3,cpd_I_dm_O_distance3, marker='.')
cpd_i_dm_aff_o = plt.xlabel('Compound I DM affinity / kcal mol $^{-1}$')
cpd_i_dm_aff_o = plt.ylabel('Closest distance from O / $\mathring{A}$')
plt.legend()

print(stats.pearsonr(cpd_I_dm_affinity1, cpd_I_dm_O_distance1))
print(stats.pearsonr(cpd_I_dm_affinity2, cpd_I_dm_O_distance2))
print(stats.pearsonr(cpd_I_dm_affinity3, cpd_I_dm_O_distance3))
```
