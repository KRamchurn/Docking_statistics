```python
#Import all relevant/useful modules

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
from pandas import ExcelFile
import scipy
from scipy import stats
import sys

#Read in Excel file and read desired columns, drop missing values

desired_cols =list(range(0,12))
df = pd.read_excel('data.xlsx', sheet_name = 'Rigid', usecols=desired_cols)
df.dropna(inplace=True)

#Assign columns to variables

wt_kd = df.iloc[:,0]
dm_kd = df.iloc[:,1]
affinity_wt = df.iloc[:,2]
affinity_dm = df.iloc[:,3]
affinity_cpd_I_wt = df.iloc[:,4]
affinity_cpd_I_dm = df.iloc[:,5]
wt_distance = df.iloc[:,6]
dm_distance = df.iloc[:,7]
cpd_I_WT_O_distance = df.iloc[:,8]
cpd_I_DM_O_distance = df.iloc[:,9]
cpd_I_WT_Fe_distance = df.iloc[:,10]
cpd_I_DM_Fe_distance = df.iloc[:,11]
```

```python
# WT KD v Affinity

kd_aff1 = plt.plot(wt_kd,affinity_wt, marker='.', linestyle='none')
kd_aff1 = plt.xlabel('WT K$_{D}$')
kd_aff1 = plt.ylabel('Affinity')

stats.pearsonr(wt_kd, affinity_wt)
```

```python
# DM KD v Affinity

kd_aff2 = plt.plot(dm_kd, affinity_dm,marker='.', linestyle='none')
kd_aff2 = plt.xlabel('DM K$_{D}$')
kd_aff2 = plt.ylabel('Affinity ')

stats.pearsonr(dm_kd, affinity_dm)
```

```python
# WT KD v CPD I WT Affinity

kd_aff3 = plt.plot(wt_kd, affinity_cpd_I_wt, marker='.', linestyle='none')
kd_aff3 = plt.xlabel('WT K$_{D}$')
kd_aff3 = plt.ylabel('Compound I WT Affinity')

stats.pearsonr(wt_kd, affinity_cpd_I_wt)
```

```python
# DM KD v CPD I DM Affinity

kd_aff4 = plt.plot(dm_kd, affinity_cpd_I_dm, marker='.', linestyle='none')
kd_aff4 = plt.xlabel('DM K$_{D}$')
kd_aff4 = plt.ylabel('Compound I DM Affinity')

stats.pearsonr(dm_kd, affinity_cpd_I_dm)
```

```python
# WT KD v Closest atom

kd_wt_dist = plt.plot(wt_kd,wt_distance, marker ='.', linestyle='none')
kd_wt_dist = plt.xlabel('WT K$_{D}$')
kd_wt_dist = plt.ylabel('Distance')

stats.pearsonr(wt_kd, wt_distance)
```

```python
# WT KD v CPD I WT Closest atom from O 

kd_cpd_I_wt_dist = plt.plot(wt_kd,cpd_I_WT_O_distance, marker ='.', linestyle='none')
kd_cpd_I_wt_dist = plt.xlabel('WT K$_{D}$')
kd_cpd_I_wt_dist = plt.ylabel('Distance from Compound I WT O')

stats.pearsonr(wt_kd, cpd_I_WT_O_distance)
```

```python
# WT KD v CPD I WT Closest atom from Fe 

kd_cpd_I_wt_dist_fe = plt.plot(wt_kd,cpd_I_WT_Fe_distance, marker ='.', linestyle='none')
kd_cpd_I_wt_dist_fe = plt.xlabel('WT K$_{D}$')
kd_cpd_I_wt_dist_fe = plt.ylabel('Distance from Compound I WT Fe')

stats.pearsonr(wt_kd, cpd_I_WT_Fe_distance)
```

```python
# DM KD v CPD I DM Closest atom from Fe

kd_cpd_I_dm_dist = plt.plot(dm_kd,cpd_I_DM_O_distance, marker ='.', linestyle='none')
kd_cpd_I_dm_dist = plt.xlabel('DM K$_{D}$')
kd_cpd_I_dm_dist = plt.ylabel('Distance from Compound I DM O')

stats.pearsonr(dm_kd, cpd_I_DM_O_distance)
```

```python
# DM KD v CPD I DM Closest atom from O 

kd_cpd_I_dm_dist_fe = plt.plot(dm_kd,cpd_I_DM_Fe_distance, marker ='.', linestyle='none')
kd_cpd_I_dm_dist_fe = plt.xlabel('DM K$_{D}$')
kd_cpd_I_dm_dist_fe = plt.ylabel('Distance from Compound I DM Fe')

stats.pearsonr(dm_kd, cpd_I_DM_Fe_distance)
```

```python
# DM KD v Distance

kd_dm_dist = plt.plot(dm_kd,dm_distance, marker ='.', linestyle='none')
kd_dm_dist = plt.xlabel('DM K$_{D}$')
kd_dm_dist = plt.ylabel('Distance to closest ligand atom')

stats.pearsonr(dm_kd, dm_distance)
```

```python
# WT Affinity v distance

wt_aff_dist = plt.plot(affinity_wt, wt_distance, marker='.', linestyle='none')
wt_aff_dist = plt.xlabel('WT affinity')
wt_aff_dist = plt.ylabel('Distance to closest ligand atom')

stats.pearsonr(affinity_wt, wt_distance)
```

```python
#DM Affinity v Distance

dm_aff_dist = plt.plot(affinity_dm, dm_distance, marker='.', linestyle='none')
dm_aff_dist = plt.xlabel('DM affinity')
dm_aff_dist = plt.ylabel('Distance to closest atom')

stats.pearsonr(affinity_dm, dm_distance)
```

```python
#CPD I DM Affinity v CPD I DM Distance to Fe

cpd_i_dm_aff_fe = plt.scatter(affinity_cpd_I_dm,cpd_I_DM_Fe_distance, marker='.')
cpd_i_dm_aff_fe = plt.xlabel('Compound I DM affinity')
cpd_i_dm_aff_fe = plt.ylabel('Closest distance from central Iron')

stats.pearsonr(affinity_cpd_I_dm, cpd_I_DM_Fe_distance)
```

```python
#CPD I DM Affinity v CPD I DM Distance to O

cpd_i_dm_aff_o = plt.scatter(affinity_cpd_I_dm,cpd_I_DM_O_distance, marker='.')
cpd_i_dm_aff_o = plt.xlabel('Compound I DM affinity')
cpd_i_dm_aff_o = plt.ylabel('Closest distance from central Oxygen')

stats.pearsonr(affinity_cpd_I_dm, cpd_I_DM_O_distance)
```

```python
#CPD I WT Affinity v CPD I WT Distance to Fe

cpd_i_wt_aff_fe = plt.scatter(affinity_cpd_I_wt,cpd_I_WT_Fe_distance, marker='.')
cpd_i_wt_aff_fe = plt.xlabel('Compound I WT affinity')
cpd_i_wt_aff_fe = plt.ylabel('Closest distance from central Iron')

stats.pearsonr(affinity_cpd_I_wt, cpd_I_WT_Fe_distance)
```

```python
#CPD I WT Affinity v CPD I WT Distance to O

cpd_i_wt_aff_o = plt.scatter(affinity_cpd_I_wt,cpd_I_WT_O_distance, marker='.')
cpd_i_wt_aff_o = plt.xlabel('Compound I WT affinity')
cpd_i_wt_aff_o = plt.ylabel('Closest distance from central Oxygen')

stats.pearsonr(affinity_cpd_I_wt, cpd_I_WT_O_distance)
```
